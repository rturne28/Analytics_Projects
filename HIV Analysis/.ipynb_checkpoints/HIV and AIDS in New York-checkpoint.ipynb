# HIV Infections in New York City
The following is an exploratory analysis of HIV infections in New York City boroughs from 2011 to 2015. This includes
* an [investigation](#infections) of HIV infections and AIDS diagnoses for all males and females from 2011 to 2015, 
* a [breakdown](#deaths) of deaths in each borough for males and females from 2011 to 2015,
* a [depiction](#suppression) of the trend in the percent of people living with diagnosed HIV infection (PLWDHI) with viral suppression in 2011 and 2015, and
* [chi squared tests](#tests) to test the null hypotheses that 
    1. the distributions of HIV diagnoses of both 50-59 yr old males and 20-29 yr old males across the four most populous boroughs are the same
    2. the HIV diagnoses in males in the Manhattan borough across all age groups belongs to the same distribution as the HIV diagnoses in males by age group for all boroughs.

This data was pulled from NYC's OpenData at https://data.cityofnewyork.us/Health/DOHMH-HIV-AIDS-Annual-Report/fju2-rdad .

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
hiv = pd.read_csv('./DOHMH_HIV_AIDS_Annual_Report.csv')
```

## Distribution of Deaths by Borough <a name='deaths'></a>

Preliminary analysis of the data below shows that, generally, the sum of the deaths for each borough for each year does not match the number of deaths given for "All" boroughs. However, the note given on the NYC OpenData page mentions that city-wide totals may include cases assigned to NYC with an unknown borough. So, the sums of the data for all boroughs should be less than the data given in the "All" fields for the boroughs, which is what is seen.

The pie charts that follow use the sums of the borough totals, so the percentages seen in the charts are correct.

```python
# Query the dataframe for males and females separately where the fields show statistics for all age groups for each borough
hiv_males=hiv.loc[(hiv['Borough']!='All') & (hiv['UHF']=='All') & (hiv['Gender']=='Male') & (hiv['Age']=='All')\
                  & (hiv['Race']=='All')]
hiv_females=hiv.loc[(hiv['Borough']!='All') & (hiv['UHF']=='All') & (hiv['Gender']=='Female') & (hiv['Age']=='All')\
                    & (hiv['Race']=='All')]
hiv_males
```

```python
hiv_females.loc[hiv_females['Year']==2015]['Deaths']
```

```python
# Create the 10 pie charts and format the labels
num_unique_years = len(pd.unique(hiv_males['Year']))
fig2, axes2 = plt.subplots(nrows=num_unique_years, ncols=2, figsize=(7.5,20))
for i in range(num_unique_years):
    year = pd.unique(hiv_males['Year'])[i]
    hiv_males.loc[hiv_males['Year']==year].reset_index(drop=True).plot.pie(y='Deaths', ax=axes2[i,0], labels=hiv_males['Borough'],
                                                                           ylabel='', xticks=[], legend=None, autopct='%1.0f%%')
    hiv_females.loc[hiv_females['Year']==year].reset_index(drop=True).plot.pie(y='Deaths', ax=axes2[i,1],
                                                                               labels=hiv_females['Borough'],
                                                                               ylabel='', 
                                                                               legend=None,
                                                                               autopct='%1.0f%%')
    axes2[i,0].axis('equal')
    axes2[i,1].axis('equal')
    axes2[i,0].set_ylabel(year, labelpad=80, fontsize=22, rotation='horizontal')
axes2[0,0].set_xlabel('Male', fontsize=22)
axes2[0,1].set_xlabel('Female', fontsize=22)
axes2[0,0].xaxis.set_label_position('top')
axes2[0,1].xaxis.set_label_position('top')
# fig2.legend(labels=pd.unique(hiv_males['Borough']))

fig2.subplots_adjust(wspace=0.75)
plt.show()
```

The pie charts for males show that, generally, there is no change in the distribution of HIV deaths across the NYC boroughs from 2011 to 2015. But, as the NYC OpenData page details, these deaths are not directly attributed to HIV infection but are **deaths from any cause** among people with HIV/AIDS. This statistic is also heavily dependent on population trends for each borough, as an influx of people with or without HIV/AIDS to one particular borough might increase the fraction of deaths from HIV/AIDS in that borough if the distribution of HIV/AIDS among this influx of people follows the same distribution of HIV/AIDS in the people in the other boroughs.

The pie charts of deaths in females with HIV/AIDS shows a little more change in the distribution over the years, namely 34% of deaths occured in the Bronx in 2011 vs 41% of deaths in the Bronx in 2015. This may be attributed to the smaller sample size analyzed when compared to the male cohort.

## Viral Suppression among PLWDHI <a name='suppression'></a>

The goal of this analysis is to show the distribution of people living with diagnosed HIV infection (PLWDHI) that have attained viral suppression. From the NYC OpenData page, viral suppression is defined as being attained when at least one viral load test during the calendar year shows an HIV viral load value of $\leq$ 200 copies/mL.

```python
# Query for all years but keep the breakdown of results for each age group
hiv_viral=hiv.loc[(hiv['Borough']=='All') & (hiv['UHF']=='All') & (hiv['Age']!='All') & (hiv['Race']=='All')]
hiv_viral
```

```python
fig3, axes3 = plt.subplots(nrows=2, sharex=True, figsize=(5,7))
hiv_viral_male=hiv_viral.loc[(hiv_viral['Year']==2011) & (hiv_viral['Gender']=='Male')]
hiv_viral_female=hiv_viral.loc[(hiv_viral['Year']==2011) & (hiv_viral['Gender']=='Female')]

hiv_viral_male_15=hiv_viral.loc[(hiv_viral['Year']==2015) & (hiv_viral['Gender']=='Male')]
hiv_viral_female_15=hiv_viral.loc[(hiv_viral['Year']==2015) & (hiv_viral['Gender']=='Female')]

hiv_viral_male.plot.bar(x='Age', y='% viral suppression', ax=axes3[0], position=0.25, color='blue', ylabel='2011', label='Male')
hiv_viral_female.plot.bar(x='Age', y='% viral suppression', ax=axes3[0], color='magenta', ylabel='2011', label='Female')
hiv_viral_male_15.plot.bar(x='Age', y='% viral suppression', ax=axes3[1], position=0.25, color='blue', ylabel='2015', label='Male')
hiv_viral_female_15.plot.bar(x='Age', y='% viral suppression', ax=axes3[1], color='magenta', ylabel='2015', label='Female')

fig3.suptitle('Percent Viral Suppression', fontsize=22)
axes3[0].set_ylabel('2011', fontsize=22, labelpad=40, rotation='horizontal')
axes3[1].set_ylabel('2015', fontsize=22, labelpad=40, rotation='horizontal')
axes3[1].set_xlabel('Age', fontsize=22, labelpad=10)
plt.show()
```

The bar charts show an encouraging trend that the fraction of PLWDHI that have attained viral suppression increased for both males and females in all age groups from 2011 to 2015. The most striking changes are 
1. the increase of males with viral suppression in the 13-19 yr age group from about 50% to almost 70% and 
2. the increase of both males and females with viral suppression in the 20-29 yr age group from about 45-55% to both around the 65-75% range.

A few possible explanations for these improvements:
* Viral suppressant medications may have been approved for use in teenagers after 2011
* Suppressant medications may have become more accessible to all, particularly for those in the 20-29 yr age group
* Knowledge of HIV infection and the consequences has prompted PLWDHI to seek treatment.

## Chi-Squared Hypothesis Tests on Distributions over Age Groups <a name='tests'></a>

The following analyses utilize the Pearson's Chi-Squared tests to determine
1. if the distributions of HIV diagnoses of both 50-59 yr old males and 20-29 yr old males across the four most populous boroughs in 2015 are the same, and
2. if the HIV diagnosis data in males from the Manhattan borough is an appropriate sample of all of the male diagnosis data in NYC in 2015.

In the first test, the diagnosis data from Staten Island is omitted, as this data suffers from low sample size.
Both tests investigate the __male__ HIV diagnosis data in __2015__, as these subsets of the entire table provide the highest counts.

```python
# Case 1
hiv_20_29=hiv.loc[(hiv['Borough']!='All') & (hiv['Borough']!='Staten Island') & (hiv['UHF']=='All') & (hiv['Age']=='20 - 29')\
                  & (hiv['Race']=='All')  & (hiv['Year']==2015) & (hiv['Gender']=='Male')]
hiv_50_59=hiv.loc[(hiv['Borough']!='All') & (hiv['Borough']!='Staten Island') & (hiv['UHF']=='All') & (hiv['Age']=='50 - 59') & (hiv['Race']=='All')\
                  & (hiv['Year']==2015) & (hiv['Gender']=='Male')]

_20_29 = hiv_20_29['HIV diagnoses']
obs = hiv_50_59['HIV diagnoses']
_20_29_ratios = _20_29 / _20_29.sum()
exp = _20_29_ratios*(obs.sum())
print(obs)
print(_20_29)
stats.chisquare(f_obs = obs, f_exp = exp)
```

In this case, we are somewhat arbitrarily assuming that the expected distribution of HIV diagnoses in males across the four most populous boroughs in 2015 is given by the counts in the 20-29 yr old age group. This assumption was made in an attempt to produce a test statistic high enough at a 95% confidence to reject the null hypothesis that the two distributions of HIV diagnoses are the same. 

However, the resultant p-value shows that the probability of observing the 2.84 test statistic or higher for the observed diagnosis distribution in the 50-59 yr old age group is 41.6% if the expected distribution is given by the HIV diagnoses in the 20-29 yr old age group. Because this is greater than 1-0.95=0.05, the null hypothesis cannot be rejected and we assume the two sets of HIV diagnosis data belong to the same distribution.

```python
# Case 2
hiv_manhattan=hiv.loc[(hiv['Borough'] == 'Manhattan') & (hiv['UHF']=='All') & (hiv['Age']!='All') \
                  & (hiv['Race']=='All')  & (hiv['Year']==2015) & (hiv['Gender']=='Male')]
hiv_all=hiv.loc[(hiv['Borough'] == 'All') & (hiv['UHF']=='All') & (hiv['Age']!='All') \
                  & (hiv['Race']=='All')  & (hiv['Year']==2015) & (hiv['Gender']=='Male')]
obs = hiv_manhattan['HIV diagnoses']
all_boroughs = hiv_all['HIV diagnoses']
ratio = all_boroughs / all_boroughs.sum()
exp = obs.sum()*ratio
print(obs)
print(all_boroughs)
stats.chisquare(f_obs = obs, f_exp = exp)
```

For this case, the expected male HIV diagnosis data is taken to be that from the entirety of NYC for 2015; the observed data is from the Manhattan borough only. The expectation is failure to reject the null hypothesis that the Manhattan cohort is an acceptable sample of all HIV diagnoses in males in NYC, as the Manhattan population is a sizeable fraction of the NYC population. While this is true, the counts in the 13-19 and 60+ yr age groups are dangerously low to exectute the chi squared test with confidence.

If the counts are deemed acceptable to carry out the chi squared test, then the null hypothesis indeed cannot be rejected at a 95% confidence, as the odds of seeing a test statistic of 8.09 or higher with 5 degrees of freedom is about 15%, which is higher than the 5% p-value needed for rejection.
